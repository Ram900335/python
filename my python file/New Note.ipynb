```python
def Reverse(N):
    if(N<10):
        return N
    v = 0;
    Iteration = math.floor(math.log10(N)+1)
    for x in range(Iteration):
        d = N % 10;
        v = v*10+d
        N = N//10
    return v
```

```python
def Isprime(Number):
    if Number == 2:
        return True
    for i in range(2,math.ceil(math.sqrt(Number)+1)):
        if(Number%i == 0):
            return False
    return True;
```
