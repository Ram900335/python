```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.externals import joblib

#load the file
df=pd.read_csv("auto_mpg.csv")
del df['origin']
del df['weight']
del df['displacement']
del df['acceleration']

#one-hot encoded data
a_df=pd.get_dummies(df, columns=['car name'])

#remove the mpg the data
del a_df['mpg']

#create the x and y arrays
x=a_df.as_matrix()
y=df['mpg'].as_matrix()

x,y
```

```python
a_df.to_csv('samplex.csv')
df['mpg'].to_csv('sampley.csv')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=30)
x_train,x_test,y_train,y_test
```

```python
#fit regression model
m=ensemble.GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.1,
    max_depth=6,
    min_samples_leaf=9,
    max_features=0.1,
    loss='huber',
    random_state=0,
)
m.fit(x_train,y_train)
```

```python
#save the trained model to a file so wee can use it in other programs
joblib.dump(m,'m/trained_auto_mpg_model.pkl')
```

```python
#find the error rate on the training set
train=mean_absolute_error(y_train, m.predict(x_train))
print("taining set mean absolute error:%4f" % train)
#find the error rate on the test set
test=mean_absolute_error(y_test, m.predict(x_test))
print("test set mean absolute error:%4f"%test)
```
