```python
print("Row1:")
for x in range(3):
    for y in range(3):
        if x==0:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("Row2:")
for x in range(3):
    for y in range(3):
        if x==1:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("Row3:")   
for x in range(3):
    for y in range(3):
        if x==2:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
    
print("col1:")
for x in range(3):
    for y in range(3):
        if y==0:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("col2:")
for x in range(3):
    for y in range(3):
        if y==1:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("col3:")   
for x in range(3):
    for y in range(3):
        if y==2:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("Diag1")
for x in range(3):
    for y in range(3):
        if y==x:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
print("Diag2")
for x in range(3):
    for y in range(3):
        if x==0 and y==2:
            print('*',end=' ')
        elif x==2 and y==0:
            print('*',end=' ')
        elif x==1 and y==1:
            print('*',end=' ')
        else:
            print('-',end=' ')
    print('\n')
```
