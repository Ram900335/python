```python
import pandas as pd
df=pd.read_csv("australian.csv",names=['a1','a2','a3','a4','a5','a6','a7','a8','a9','a10','a11','a12','a13','a14','a15'])
df=df.drop(['a2','a3','a7','a10','a13','a14'],axis=1)
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
#df=pd.read_csv("german.csv", error_bad_lines=False,delimiter=';')
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
#clf=DecisionTreeClassifier(max_leaf_nodes=15)
clf=DecisionTreeClassifier()
clf=clf.fit(train.drop(['a15'],axis=1).values,train['a15'].values)
```

```python
clf
```

```python
clf.feature_importances_
```

```python
from sklearn import tree
with open("german_bank.dot","w") as f:
    f:tree.export_graphviz(clf,feature_names=train.drop(['a15'],axis=1).columns,out_file=f)
```

```python
predictions=clf.predict(test.drop(['a15'],axis=1).values)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test['a15'],predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
x_train=train.drop(['a15'],axis=1).values
y_train=train['a15'].values
x_test=test.drop(['a15'],axis=1).values
y_test=test['a15']

clf = RandomForestClassifier(n_estimators=1000,max_leaf_nodes=15)
clf=clf.fit(x_train,y_train)
predictions = clf.predict(x_test)
accuracy_score(y_test, predictions)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf=GradientBoostingClassifier()
clf=clf.fit(x_train,y_train)
predictions=clf.predict(x_test)
accuracy_score(y_test,predictions)
```
